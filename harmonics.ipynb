```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = x = np.arange(0,1,0.001)
v = np.sin(2*np.pi*x) 
y = 0.8*np.sin(2*np.pi*x - 0.4) 
xax = np.zeros(1000)

plt.plot(x,v,label = 'Voltage')
plt.plot(x,y, label = 'current')
plt.plot(x,xax)
plt.legend()
```

```python
x = x = np.arange(0,1,0.001)
v = np.sin(2*np.pi*x) 
y1 = 0.6*np.sin(2*np.pi*x - 0.4) 
y3 = 0.08*np.sin(2*np.pi*x*3 - 0.4*3)
y5 = 0.05*np.sin(2*np.pi*x*5 - 0.4*5)
y7 = 0.05*np.sin(2*np.pi*x*7 - 0.4*7)
xax = np.zeros(1000)

plt.plot(x,xax)
plt.plot(x,y1,label = 'Fundamental')
plt.plot(x,y3,label = '3rd Harmonics')
plt.plot(x,y5,label = '5th Harmonics')
plt.plot(x,y7,label = '7th Harmonics')
plt.plot(x,y1+y3+y5, label = 'Distorted Current Wave')
plt.legend()

THD_i = 100 * ((0.08**2 + 0.05**2 + 0.05**2)/0.6**2)**(0.5)
THD_i
```
